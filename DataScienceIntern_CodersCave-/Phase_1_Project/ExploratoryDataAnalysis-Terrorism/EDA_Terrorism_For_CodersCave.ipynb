Project Summary:




Exploratory Data Analysis (EDA) is a critical preliminary step in the data analysis process. It involves the following key activities:

1.Data Familiarization: Begin by loading and understanding the dataset's structure. You check the number of records (rows) and features (columns) in the dataset, as well as the data types of each feature.

2.Data Cleaning: This step involves addressing data quality issues, such as handling missing values, removing duplicates, and correcting data types. It ensures that the data is in a suitable form for analysis.

3.Data Distribution: Understanding the distribution of numerical features is important for identifying trends and patterns. You can calculate summary statistics (mean, median, standard deviation, etc.) and visualize data distributions using histograms, box plots, or density plots.

4.Data Visualization: EDA includes creating various plots and visualizations to explore relationships and patterns in the data. These visualizations can include scatter plots, bar charts, time series plots, heatmaps, and geospatial maps, among others.

5.Correlations: Investigate relationships and associations between variables in the dataset. For instance, you might explore whether there is a correlation between two numerical variables using correlation matrices or scatterplots.

6.Outliers: Identify and investigate outliers in the data. Outliers are data points that significantly deviate from the majority of the data. These could be important insights or errors that need to be addressed.

7.Reporting and Visualization: Summarize the findings, insights, and conclusions from the EDA process. Effective data visualization helps convey these key takeaways to stakeholders and decision-makers.

EDA is a crucial step in the data analysis process as it helps data analysts and scientists gain a comprehensive understanding of the dataset, discover patterns, and identify potential issues that may affect subsequent analyses or modeling efforts. It provides the foundation for making data-driven decisions and extracting valuable insights from the data.

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
data=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum() #here also have some null value like state city etc..
```

```python
data['Summary'].fillna("Not registered", inplace=True)
data['Target'].fillna("Not registered", inplace=True)
data['Motive'].fillna("Not Find", inplace=True)
data["Killed"].fillna(0,inplace=True)
data["Wounded"].fillna(0,inplace=True)
data['state'].fillna("Not registered", inplace=True)
data.dropna(subset=['latitude', 'longitude'], inplace=True)
data['city'].fillna("Not registered", inplace=True)
```

```python
data.isnull().sum()
```

```python
#@title top attacked countries


top_attacked_countries=data['Country'].value_counts()[0:100].reset_index().rename(columns={'index':'Country','Country':'Attacked'})
top_attacked_countries.style.background_gradient()
```

```python
less_attacked_countries=data['Country'].value_counts()[-10:]
less_attacked_countries
```

```python
death_data=data.groupby('Year')['Killed'].sum().reset_index().sort_values('Killed')
death_data.style.background_gradient()
```

```python
#@title Data Vizualization --> Chart 1
plt.figure(figsize=(16,9))
sns.countplot(data=data,x="Year")
plt.title('Count of terrorist activities each year')
plt.xticks(rotation=90)
plt.show()
```

```python
#@title Chart 2
plt.figure(figsize=(16,9))
sns.countplot(data=data,x="Region")
plt.title("Count of terrorist activities by region")
plt.xticks(rotation=90)
plt.show()
```

```python
#@title Chart 3
plt.figure(figsize=(16,9))
sns.lineplot(data=data,x="Year",y="Killed",estimator="sum")
plt.title('Number of people killed by terror attack')
plt.xticks(rotation=90)
plt.show()
```

```python
#@title Chart - 4
plt.figure(figsize=(16,9))
sns.countplot(data=data, x='AttackType')
plt.title('Attack types')
plt.xticks(rotation=90)
plt.show()
```

```python
#@title Chart - 5
plt.figure(figsize=(16,9))

# Get the count of each group
group_data = data[data['state'] != 'Not register']['state'].value_counts().head(10)

# Use this count for plotting
sns.barplot(x=group_data.index, y=group_data.values)

plt.title('Top 10 terrorist groups with highest no of attacks')
plt.xticks(rotation=90)
plt.xlabel('Group')
plt.ylabel('Count')
plt.show()
```
